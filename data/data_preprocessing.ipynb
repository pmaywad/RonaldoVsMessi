```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("MessiRonaldo.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df['All_Goals'] = df['Liga_Goals'] + df ['CL_Goals']
df['All_Asts'] = df['Liga_Asts'] + df ['CL_Asts']
df['All_Aps'] = df['Liga_Aps'] + df ['CL_Aps']

#Goals per Match
df['Liga_GPM'] = (df['Liga_Goals']/df['Liga_Aps']).round(2)
df['CL_GPM'] = (df['CL_Goals']/df['CL_Aps']).round(2)
df['All_GPM'] = (df['All_Goals']/df['All_Aps']).round(2)

#Assists per Match
df['Liga_APM'] = (df['Liga_Asts']/df['Liga_Aps']).round(2)
df['CL_APM'] = (df['CL_Asts']/df['CL_Aps']).round(2)
df['All_APM'] = (df['All_Asts']/df['All_Aps']).round(2)

```

```python
df.describe()
```

```python
df.to_csv("MessiRonaldo_Updates.csv")
```
